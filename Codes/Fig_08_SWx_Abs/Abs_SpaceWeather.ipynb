# تحميل (استيراد) المكتبات الأساسية
# Importing

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# تحميل البيانات Data downloading

```python
# Load the CSV file
df = pd.read_csv('Years.csv')

# Clean column names by stripping extra spaces
df.columns = df.columns.str.strip()

# Convert relevant columns to numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df['Article Count'] = pd.to_numeric(df['Article Count'], errors='coerce')

# Filter to include only data from 1974 onward
df_filtered = df[(df['Year'] >= 1974) & (df['Year'] <= 2022)].dropna(subset=['Year', 'Article Count'])

# Extract values for plotting
years = df_filtered['Year'].astype(float).values
counts = df_filtered['Article Count'].astype(float).values

```

# العرض Visualization

```python
plt.figure(figsize=(12, 6), dpi = 600)
plt.fill_between(years, counts, color='skyblue', alpha=0.7)
plt.title('NASA/ADS Publications with "Space Weather" in Abstract (1974–2022)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Publications', fontsize=14)
plt.xticks(ticks=range(1975, 2023, 5), fontsize=12)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.grid(True, which='major', linestyle='--', alpha=0.5)
plt.ylim(0, max(counts) + 10)
plt.tight_layout()
plt.show()
```
